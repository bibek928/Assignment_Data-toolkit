1. What is NumPy, and why is it widely used in Python?
  - NumPy (Numerical Python) is a powerful library used for numerical computing. It provides support for arrays, matrices, and many mathematical functions. It's widely used due to its performance, ease of use, and integration with other scientific libraries.

2. How does broadcasting work in NumPy?
  - Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by stretching the smaller array to match the shape of the larger one, without copying data.

3. What is a Pandas DataFrame?
  - A DataFrame is a 2D, mutable, tabular data structure with labeled axes (rows and columns) used for data manipulation and analysis in Pandas.

4. Explain the use of the groupby() method in Pandas.
  - The groupby() method is used to group data based on one or more columns and apply aggregate functions like sum(), mean(), etc., to each group.

5. Why is Seaborn preferred for statistical visualizations?
  - Seaborn offers a high-level interface for attractive and informative statistical graphics, making complex plots easier to create and interpret.

6.  What are the differences between NumPy arrays and Python lists?
  - NumPy arrays are faster and more efficient.
  - Arrays support vectorized operations.
  - They require all elements to be of the same data type.

7. What is a heatmap, and when should it be used?
   - A heatmap visualizes matrix-style data with color intensity. It's useful for displaying correlations or data densities.

8. What does the term “vectorized operation” mean in NumPy?
   - Vectorized operations allow element-wise operations on entire arrays without writing loops, resulting in faster execution.

9. How does Matplotlib differ from Plotly?

   - Matplotlib: Static, script-based plots.
   - Plotly: Interactive, web-based visualizations with zoom and hover features.

10. What is the significance of hierarchical indexing in Pandas?
   - It enables working with higher-dimensional data in a 2D DataFrame by using multiple index levels (MultiIndex).

11. What is the role of Seaborn’s pairplot() function?
    - It creates a matrix of scatter plots for all numerical variable combinations, useful for exploring relationships.

12. What is the purpose of the describe() function in Pandas?
   - It summarizes statistics (mean, std, min, max, etc.) of numerical columns in a DataFrame.

13. Why is handling missing data important in Pandas?
    - Missing data can distort analysis. Pandas provides tools to detect, fill, or drop missing values for cleaner datasets.

14. What are the benefits of using Plotly for data visualization?
    - Plotly offers interactivity, beautiful visuals, integration with Dash for web apps, and works well in Jupyter Notebooks.

15. How does NumPy handle multidimensional arrays?
    - NumPy uses ndarray, which can represent arrays with any number of dimensions and supports a wide range of operations.

16. What is the role of Bokeh in data visualization?
    - Bokeh is used for creating interactive, web-based plots, especially useful for large streaming datasets.

17. Explain the difference between apply() and map() in Pandas.
    - map(): Element-wise operation on Series.
    - apply(): Applies a function to each row/column of a DataFrame or Series.

18. What are some advanced features of NumPy?
    - Broadcasting
    - Masked arrays
    - Structured arrays
    - Universal functions (ufuncs)
    - Memory-mapped files

20. How does Pandas simplify time series analysis?
  - Pandas provides datetime indexing, resampling, frequency conversion,
     and easy handling of date-time data.

21. What is the role of a pivot table in Pandas?
   - It summarizes data by grouping and aggregating values, useful for multidimensional analysis like Excel pivot tables.

22. Why is NumPy’s array slicing faster than Python’s list slicing?
   - NumPy arrays use contiguous memory and vectorized C operations, while
     lists store references to objects, making slicing slower.

23. What are some common use cases for Seaborn?

  - Distribution plots (histogram, KDE)
  - Categorical plots (box, violin)
  - Heatmaps for correlation
  - Regression plots
